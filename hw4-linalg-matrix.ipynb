**Определение**

*Линейным подпространством* $V \subset \mathbb R^n$ называется такое подмножество векторов, что для любых $u, v\in V$ и любого скаляра $\lambda \in \mathbb R$:

- $u + v \in V$;
- $\lambda u \in V$.

*Аффинным подпространством* $W \subset \mathbb R^n$  множество векторов $\{u_0 + v \mid v \in  V\}$, получаемых из некоторого фиксированного вектора $u_0$ прибавлением всевозможных векторов из линейного подпространства $V$. Иными словами, аффинное подпространство — это результат сдвига линейного подпространства на некоторый фиксированный вектор.
